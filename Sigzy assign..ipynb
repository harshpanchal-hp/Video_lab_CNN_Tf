## First we will convert video to images using opencv

```python
import os
import math
import cv2
import math
count = 1
for file in parent:
    print(file)
    video = cv2.VideoCapture("./jogging/" + file)
    print(video.isOpened())
    framerate = video.get(5)
    os.makedirs("./jogging/" + "video_" + str(int(count)))
    while (video.isOpened()):
        frameId = video.get(1)
        success,image = video.read()
        if( image != None ):
            image=cv2.resize(image,(224,224), interpolation = cv2.INTER_AREA)
        if (success != True):
            break
        if (frameId % math.floor(framerate) == 0):
            filename = "./jogging/video_" + str(int(count)) + "/image_" + str(int(frameId / math.floor(framerate))+1) + ".jpg"
            print(filename)
            cv2.imwrite(filename,image)
    video.release()
    print('done')
    count+=1
```
